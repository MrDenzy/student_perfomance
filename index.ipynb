```python
# your work here 
import pandas as pd
import statsmodels.api as sm
import seaborn as sns 

df = pd.read_csv('data\Student_Performance.csv')
df.head()
```

```python
df.isna().sum()
```

```python
sns.pairplot(df)
```

```python
# Define the dependent variable (y) and independent variable (X)
X = df[['Previous Scores','Hours Studied','Sample Question Papers Practiced']]  # independent variable
y = df['Performance Index']             # dependent variable

# Add a constant to X (for the intercept term)
X = sm.add_constant(X)

# Fit the regression model. Actual Training Step
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())

# Predict sales using the model
df['Student_Perfomance'] = model.predict(X)

# Display first few rows
df.head()
```
